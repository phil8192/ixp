# internet vs road traffic features

```r
workday <- function(x) ! (x == "Saturday" | x == "Sunday")
```

```r
internet <- read.csv("bandwidth_features_london.csv")
internet$X <- as.Date(as.character(internet$X))
internet <- internet[internet$X <= "2018-10-31", ]
internet <- internet[workday(weekdays(internet$X)), ]
tail(internet)
```

```r
colnames(internet)
```

```r
road <- read.csv("sa_road_features_london.csv")
road <- road[workday(road$day), ]
tail(road)
```

```r
colnames(road)
```

```r
nrow(internet) == nrow(road) && all(as.character(internet$X) == as.character(road$date))
```

```r
cors <- cor(internet[, -1], road[, 1:(ncol(road)-3)], use="pairwise.complete.obs")
cors
heatmap(cors)
```

```r
my_melt <- function(x) {
  row_names <- rep(rownames(x), ncol(x))
  col_names <- unlist(lapply(colnames(x), function(cn) rep(cn, nrow(x))))
  melt_vals <- as.vector(x)
  data.frame(row=row_names, col=col_names, val=melt_vals)
}
```

```r
#options(repr.matrix.max.rows=10)
melted <- my_melt(cors)
melted <- melted[order(-melted$val), ]



melted
```
