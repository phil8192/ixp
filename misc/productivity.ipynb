```r
x <- read.csv("../../data/econ/labour_productivity_by_industry_division_output_per_hour_sa.csv", stringsAsFactors=F)
sic.names <- x[1, ]
x <- x[-1, ]
head(x)
```

```r
options(repr.matrix.max.rows=100)
lala <- t(sic.names)
lala <- lala[-1, ]
lala <- data.frame(x=1+(1:length(lala)), lala)
lala
```

```r
tech <- tail(x[, c(33:38, 46, 48)], 12)
tech <- apply(tech, 2, as.numeric)
tech
```

```r
tech.mean <- rowMeans(tech)
tech.mean
```

```r
options(repr.plot.width=10, repr.plot.height=5)
plot(tech.mean, type="b", bty="n")
```

```r
internet <- read.csv("../../traffic_comp/bandwidth_features_london.csv", stringsAsFactors=F)
internet$X <- as.Date(internet$X)
internet$year <- as.numeric(strftime(internet$X, "%Y"))
internet$month <- as.numeric(strftime(internet$X, "%m"))
internet$day <- weekdays(internet$X)
internet$day.n <- as.numeric(strftime(internet$X, "%w"))

# remove weekends
is_weekday <- function(x) !(x == "Saturday" | x == "Sunday")
internet <- internet[is_weekday(internet$day), ]

head(internet)
colnames(internet)
```

raw daily bandwidth (raw)

```r
q <- data.frame(y=rep(2016:2018, each=12),  q=rep(1:4, each=3), bw=as.vector(with(internet, tapply(raw.Mean, list(month, year), mean))))
q <- as.vector(with(q, tapply(bw, list(q, y), mean)))
q
```

```r
plot(q, type="b", bty="n")
par(new=T)
plot(tech.mean, type="l", bty="n", lty=2)
```

```r
cor.test(diff(tech.mean), diff(q))
```

```r
ccf(diff(tech.mean), diff(q))
```
